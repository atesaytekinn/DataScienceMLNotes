### Concatenation

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
np.concatenate([x,y])
```

```python
a = np.array([[1,2,3],[4,5,6]])
np.concatenate([a,a,a]) # iki boyutlu arrayleri satır bazında birleştirme
```

```python
np.concatenate([a,a], axis = 1) # iki boyurlu arrayleri sütun bazında birleştirme
```

Arrayleri birleştirmek için kullanılır.

### Splitting

```python
x = np.array([1,2,3,99,99,3,2,1])
np.split(x, [3,5]) # 3'e kadar ve 5'e kadar ayır
```

```python
a,b,c = np.split(x, [3,5]) # ayırdıklarımızı atamak için
c
```

```python
m = np.arange(16).reshape(4,4)
np.vsplit(m,[2]) # 2. satıra kadarını ayır
```

```python
np.hsplit(m, [2]) # 2. sütuna kadarını ayır
```

```python
Yapısını bildiğimiz arrayleri parçalamak için kullanılır.
```

### Sorting

```python
v = np.array([2,1,4,3,5])
np.sort(v)
```

```python
v # hala aynı
```

```python
v.sort()
v # şimdi yapı değişti
```

```python
m = np.random.normal(20,5,(3,3)) # 2 boyutta sıralama
np.sort(m, axis = 1) # her bir satırı kendi içinde küçükten büyüğe sıralar
```

```python
np.sort(m, axis = 0) # her bir sütunu kendi içinde sıralar
```
