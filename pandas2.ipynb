### Gözlem ve Değişken Seçimi: loc & iloc

```python
import numpy as np
m = np.random.randint(1,30,size=(10,3))
df = pd.DataFrame(m, columns = ["var1","var2","var3"])
df
```

**loc** => tanımlandığı şekli ile seçim yapmak için kullanılır.

```python
df.loc[0:3] # !! 3. index dahil !!
```

**!!** 3. index dahil **!!**

**iloc** => alışık olduğumuz indexleme mantığı ile seçim yapar.

```python
df.iloc[0:3]
```

```python
df.iloc[:3,:2]
```

```python
df.loc[0:3,"var3"]
```

```python
## df.iloc[0:3,"var3"] hata verir.
df.iloc[0:3][["var3"]]
```

### İleri Toplulaştırma İşlemleri

#### aggregate

```python
df = pd.DataFrame({'group':["A","B","C","A","B","C"], 
                   'data1':[10,11,52,23,43,55],
                   'data2':[100,253,333,262,111,969]})
df
```

İstediğimiz aggregation fonksiyonlarını seçip beraber uygulamak için kullanılır.

```python
df.groupby("group").aggregate(["min",np.median,max]) # istediğimiz agg. funcları seçtik
# pandasta tanımlı olan min ve maxı farklı, dışarıdan aldığım median foknisyonunu farklı yazdım!!
# min ve maxı tırnak içinde yazıp yazmamamız önemli değil!!
```

```python
df.groupby("group").aggregate({"data1":"min", "data2":max})
# iki değişken için farklı istatistikleri hesapladı - sözlük yapısı ile
```

#### filter

```python
df.groupby("group").std()
```

```python
def filter_func(x):
    return x["data1"].std() > 9
df.groupby("group").filter(filter_func)
```

```python
(df.groupby("group").filter(filter_func)).groupby("group")["data1"].std() # I did!
```

#### transform

```python
df["data1"]*9
```

```python
df1 = df.iloc[:,1:] # group özelliğinden kurtulduk.
df1
```

```python
df1.transform(lambda x: x-x.mean())
```

### Pivot Tablolar

Veri setleri üzerinde bazı satır ve sütun işlemleri ypaarak veri setini amaca uygun hale getirmek için kullanılan yapılardır. Çok boyutlu groupby() olarak düşünülebilir.

```python
t = sns.load_dataset("titanic")
t.head()
```

```python
t.groupby("sex")["survived"].mean()
```

```python
t.groupby("sex")[["survived"]].mean() # dataFrame olarak gelir iki [] yaparsak
```

```python
t.groupby(["sex","class"])[["survived"]].aggregate("mean")
```

```python
t.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

```python
t.pivot_table("survived", index="sex", columns = "class") # kısa yol
```

**ex:** Yaşları da kategorileştirerek pivot table'a ekle.

```python
age = pd.cut(t["age"], [0,18,90]) # yaşları verilen değerleri göre böl
age.head(10)
```

```python
t.pivot_table("survived", ["sex",age], "class")
# age'i tırnak içine alma ki bizim yarattığımız age'i dışarıdan alsın.
```
